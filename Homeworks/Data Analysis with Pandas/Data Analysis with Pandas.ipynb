```python
client=pd.read_csv('client_list.csv')
```

```python
customer=pd.read_csv('customer_id_list.csv')
```

```python
client.head()
```

```python
customer.head()
```

```python
client.columns=client.columns.str.upper()
```

```python
client.head()
```

```python
clients = pd.merge(left=customer,right=client,how='left',on=['FIRST_NAME','LAST_NAME','AGE','GENDER'])
```

```python
clients
```

```python
purchase=pd.read_csv('purchase_log.csv')
```

```python
purchase
```

```python
detailed_sales=pd.merge(left=purchase,right=clients,how='inner',on=['CUSTOMER_ID'])
```

```python
detailed_sales
```

```python
spenders = detailed_sales.groupby(['CUSTOMER_ID','FIRST_NAME','LAST_NAME','AGE'], as_index=False)['PRICE'].sum()
```

```python
spenders[spenders.PRICE==spenders.PRICE.max()]
```
